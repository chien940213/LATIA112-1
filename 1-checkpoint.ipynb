```python
import pandas as pd
import io
#open the file in binary mode and read it
with open('僑生、港澳生數及其在學比率.csv','rb') as file:
    content = file.read().decode('utf-8', errors='ignore')

#create a dataframe from the content
df = pd.read_csv(io.StringIO(content))
# 讀取CSV檔案
data = pd.read_csv('僑生、港澳生數及其在學比率.csv')  # 替換為你的檔案路徑

import matplotlib.pyplot as plt
import matplotlib
matplotlib.rc('font', family='Microsoft JhengHei')
```

```python
df
```

```python
type_list = []
for i in df['學校名稱']:
    if ('國立' in i) or ('市立' in i):
        type_list.append('國立')
    else:
        type_list.append('私立')

df['公私立'] = type_list # 將 Dataframe 新增「公私立」column
df1 = df.drop_duplicates('學校統計處代碼')
count = df1['公私立'].value_counts()

print(f"本資料集共收集了 {len(df1)} 所學校，其中公立：{count['國立']} 所；私立：{count['私立']} 所。")
categories = ['國立','私立']
typecount=[count['國立'],count['私立']]
colors = ['skyblue', 'lightcoral', 'teal']
plt.bar(categories,typecount, color=colors)
plt.ylabel('學生人數')
plt.title('106-110學年度一般大學、技專校院的僑生、港澳生學生數量')
plt.show()
```

```python
students1 = data[data['學校類別'] == '一般大學']
students1_count = students1['僑生小計'].sum()
students2 = data[data['學校類別'] == '技專校院']
students2_count = students2['僑生小計'].sum()
students3 = data[data['學校類別'] == '一般大學']
students3_count = students3['港澳生小計'].sum()
students4 = data[data['學校類別'] == '技專校院']
students4_count = students4['港澳生小計'].sum()

print("就讀一般大學的僑生學生數量：", students1_count)
print("就讀技專校院的僑生學生數量：", students2_count)
print("就讀一般大學的港澳生學生數量：", students3_count)
print("就讀技專校院的港澳生學生數量：", students4_count)

# 數據
categories = ['一般大學僑生人數', '技專校院僑生人數', '一般大學港澳生人數','技專校院港澳生人數']
student_counts = [students1_count, students2_count, students3_count, students4_count]
colors = ['skyblue', 'lightcoral', 'teal']

# 創建條形圖
plt.bar(categories, student_counts, color=colors)
plt.ylabel('學生人數')
plt.title('106-110學年度一般大學、技專校院的僑生、港澳生學生數量')
# 標示每個條形的數量
for i in range(len(student_counts)):
    plt.text(categories[i], student_counts[i]-10000, str(student_counts[i]), ha='center', va='bottom')

plt.show()
```

```python
# 取得男生數量和女生數量
male1 = '僑生男'
selected_male1 = data[male1]
column_sum_male1 = selected_male1.sum()
male2 = '港澳生男'
selected_male2 = data[male2]
column_sum_male2 = selected_male2.sum()
female1 = '僑生女'
selected_female1 = data[female1]
column_sum_female1 = selected_female1.sum()
female2 = '港澳生女'
selected_female2 = data[female2]
column_sum_female2 = selected_female2.sum()
column_sum_male =column_sum_male1+column_sum_male2
column_sum_female =column_sum_female1+column_sum_female2
# 計算男女比例
total_students = column_sum_male + column_sum_female
male_percentage = (column_sum_male / total_students) * 100
female_percentage = (column_sum_female / total_students) * 100

# 將比例四捨五入到小數點第二位
male_percentage = round(male_percentage, 2)
female_percentage = round(female_percentage, 2)

print("男生數量：", column_sum_male)
print("女生數量：", column_sum_female)
print("男生比例：", male_percentage, "%")
print("女生比例：", female_percentage, "%")

#數據
labels = ['男生', '女生']
sizes = [column_sum_male, column_sum_female]
colors = ['lightblue', 'pink']
explode = (0, 0.1)  #

#繪製圓餅圖
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # 使圓餅圖為等比例
plt.title('所有學生的男女比例')
plt.show()
```

```python

# 計算台師大的人數總計
taiwan_normal_university = data[data['學校名稱'] == '國立臺灣師範大學']
total_students_count1 = taiwan_normal_university['僑生小計'].sum()
total_students_count2 = taiwan_normal_university['港澳生小計'].sum()
total_students_count =total_students_count1+total_students_count2 

# 計算台大學生的人數
ntu = data[data['學校名稱'] == '國立臺灣大學']
total_students_count_ntu1 = ntu['僑生小計'].sum()
total_students_count_ntu2 = ntu['港澳生小計'].sum()
total_students_count_ntu = total_students_count_ntu1+total_students_count_ntu2

# 計算台科大學生的人數
ntust = data[data['學校名稱'] == '國立臺灣科技大學']
total_students_count_ntust1 = ntust['僑生小計'].sum()
total_students_count_ntust2 = ntust['港澳生小計'].sum()
total_students_count_ntust = total_students_count_ntust1+total_students_count_ntust2

print("國立台灣大學學生數量：", total_students_count_ntu)
print("國立科技大學學生數量：", total_students_count_ntust)
print("國立師範大學學生數量：", total_students_count)

# 三所學校的名稱
schools = ['國立台灣大學', '國立台灣科技大學', '國立台灣師範大學']

# 對應的學生人數
students_counts = [total_students_count_ntu, total_students_count_ntust, total_students_count]

# 創建一個長條圖
plt.figure(figsize=(10, 6))  # 設定圖表尺寸

# 繪製長條圖
plt.bar(schools, students_counts, color=['skyblue', 'lightcoral', 'teal'])

# 加入數值標籤
for i, count in enumerate(students_counts):
    plt.text(schools[i], count - 1000, str(count), ha='center', va='bottom', fontsize=12)

# 加入標題和軸標籤
plt.title('三所學校學生人數')
plt.xlabel('學校名稱')
plt.ylabel('學生人數')

# 顯示長條圖
plt.show()
```
